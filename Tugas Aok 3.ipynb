# EXPLORATORY DATA ANALYSIS

## Exploratory Data Analysis memungkinkan analyst memahami isi data yang digunakan, mulai dari distribusi, frekuensi, korelasi dan lainnya

```python
# Memasukan library yang akan digunakan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('data1.csv')
```

```python
data.info()
```

```python
# Untuk Menampilkan 10 data pertama
data.head(10)
```

```python
# Untuk menampilkan 10 data terakhir
data.tail(10)
```

```python
# Menampilkan Dimensi data
data.shape
```

```python
# Melihat Fitur data yang disediakan
data.columns
```

```python
# Deskripsi data
data.describe()
```

```python
data.corr()
```

```python
correlations = data.corr(method='kendall')
correlations 
```

```python
# Correlation map
# Nilai korelasi yang tinggi ditunjukan dengan warna yang lebih terang
f,ax = plt.subplots(figsize=(20,20))
sns.heatmap(correlations, annot=True, linewidths=.5, fmt='.3f', ax=ax)
plt.show
```
